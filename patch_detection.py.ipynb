```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image
import numpy as np
import os


model_path = r"C:\Users\linsha neha\OneDrive\Documents\Desktop\dataset creation\model\disaster_classifier_model.h5"
model = load_model(model_path)


image_path = r"C:\Users\linsha neha\OneDrive\Documents\Desktop\dataset creation\test_images\test_satellite_large.jpg"
img = Image.open(image_path).convert('RGB')
width, height = img.size
patch_size = 128


labels = ["Drought", "Flood", "Normal", "Wildfire"]
results = []

for y in range(0, height - patch_size + 1, patch_size):
    for x in range(0, width - patch_size + 1, patch_size):
        patch = img.crop((x, y, x + patch_size, y + patch_size))
        patch = patch.resize((128, 128))
        patch_array = img_to_array(patch) / 255.0
        patch_array = np.expand_dims(patch_array, axis=0)
        pred = model.predict(patch_array)
        results.append(np.argmax(pred))


from collections import Counter
summary = Counter([labels[r] for r in results])
print("Patch-wise Prediction Summary:")
print(summary)
```
